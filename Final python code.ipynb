```python
import numpy as np
import pandas as pd
```

```python
#Reading data from excel file.
data=pd.read_excel("Credit Banking.xlsx")
```

```python
#Replacing the age less than 18 with the mean value of age.
mean_age = data.loc[data['Age'] >= 18, 'Age'].mean()
data.loc[data['Age'] < 18, 'Age'] = mean_age
```

```python
data
```

```python
#Customer Spend sheet
spend=pd.read_excel("Credit Banking.xlsx",sheet_name="Spend")
```

```python
#Customer Repayment sheet
repay=pd.read_excel("Credit Banking.xlsx",sheet_name="Repayment")
```

```python
#inner-join to merge spend and repayment sheet: Amount_x=Spend Amount, Amount_y=Repayment Amount
merged_data = spend.merge(repay, on=['Customer', 'Month'], how='inner')
```

```python
merged_data
```

```python
#Grouping the customers data month wise
grouped_data = merged_data.groupby(['Customer', 'Month']).agg({'Amount_x': 'sum', 'Amount_y': 'sum'}).reset_index()
```

```python
grouped_data
```

```python
grouped_data['Month'] = pd.to_datetime(grouped_data['Month'])

# Extracting the year and month from the 'Month' column
grouped_data['Year'] = grouped_data['Month'].dt.year
grouped_data['Month'] = grouped_data['Month'].dt.month

# Grouping the data by 'Customer', 'Year', and 'Month' columns
grouped_datanew2 = grouped_data.groupby(['Customer', 'Year', 'Month']).sum()
```

```python
grouped_datanew2
```

```python
groupT=grouped_datanew2
groupTask=grouped_datanew2
task8=groupT
```

```python
groupnew=groupT
```

```python
#Monthly Spend of each Customer
# Reset the index and keep the 'Customer', 'Year', 'Month', and 'Amount_x' columns
Task1 = groupnew.reset_index()[['Customer', 'Year', 'Month', 'Amount_x']]
```

```python
Task1
```

```python
#Monthly Repayment of each customer
Task2 = groupnew.reset_index()[['Customer', 'Year', 'Month', 'Amount_y']]
```

```python
Task2
```

```python
#Grouping the customer from the above data of customers to get the highest paying 10 customers
grouped_repay=repay.groupby(['Customer']).agg({'Amount':'sum'}).reset_index()
```

```python
sorted_data=grouped_repay.sort_values('Amount', ascending=False)
Task3 = sorted_data.head(10)
Task3
```

```python

grouped_spend=spend.groupby(['Customer']).agg({'Amount':'sum'}).reset_index()
```

```python
#Merging spend and customer's data 
merged_spend_customer=data.merge(grouped_spend,on=(['Customer']),how="inner")
```

```python
merged_spend_customer
```

```python
merged_spend_customer2=merged_spend_customer
```

```python
#Grouping according to segment
grouped_segment=merged_spend_customer.groupby(['Segment']).agg({'Amount':'sum'}).reset_index()
grouped_segment
```

```python
#Calculating most spending segment
sort_data=grouped_segment.sort_values('Amount', ascending=False)
Task4 = sort_data.head(1)
Task4
```

```python
# Grouping the customers in age group of 10.
age_groups = merged_spend_customer2.groupby(pd.cut(merged_spend_customer2['Age'], bins=range(0, 101, 10)))['Amount'].sum()
```

```python
#Calculating the most spending age group
highest_age_group_sorted = age_groups.sort_values(ascending=False)
highest_age_group=highest_age_group_sorted.head(1)

```

```python
highest_age_group_df = pd.DataFrame({'Age Group': [str(highest_age_group.index[0])], 'Total Amount': [highest_age_group.values[0]]})
highest_age_group_df
```

```python
#Merging Customer,Spend and Repayment sheet
merged_spend_repay_customer=merged_spend_customer.merge(grouped_repay,on=(['Customer']),how="inner")
merged_spend_repay_customer
```

```python
#Calculatind debt and Interest on debt , hence profit
merged_spend_repay_customer['Debt']=merged_spend_repay_customer['Amount_x']-merged_spend_repay_customer['Amount_y']
merged_spend_repay_customer['Debt'] = merged_spend_repay_customer['Debt'].apply(lambda x: max(0, x))  # Set negative surplus values to 0


merged_spend_repay_customer['interest'] = merged_spend_repay_customer['Debt'] * 0.029
merged_spend_repay_customer
```

```python
grouped_segment2=merged_spend_repay_customer.groupby(['Segment']).agg({'interest':'sum'}).reset_index()
```

```python
#Calculating most profitable segment
sort_data2=grouped_segment2.sort_values('interest', ascending=False)
sort_data2 = sort_data2.rename(columns={'interest': 'Profit'})
most_profit_segment = sort_data2.head(1)
most_profit_segment
```

```python
#Grouping by Category(Type)
grouped_task7 = merged_data.groupby(['Type']).agg({'Amount_x': 'sum'}).reset_index()
```

```python
grouped_task7
```

```python
grouped_task7['Type'] = grouped_task7['Type'].replace({'AIR TICKET': 'Travel', 'BUS TICKET': 'Travel', 'TRAIN TICKET': 'Travel','AUTO': 'Travel','PETRO': 'Travel'})
grouped_task7['Type'] = grouped_task7['Type'].replace({'FOOD': 'Food',})
grouped_task7['Type'] = grouped_task7['Type'].replace({'RENTAL': 'Rent',})
grouped_task7['Type'] = grouped_task7['Type'].replace({'JEWELLERY': 'Misc.', 'SANDALS': 'Misc.','CAMERA': 'Misc.','CLOTHES': 'Misc.'})
grouped_task7['Type'] = grouped_task7['Type'].replace({'MOVIE TICKET': 'Entertainment'})
grouped_task7['Type'] = grouped_task7['Type'].replace({'CAR': 'Luxury', 'BIKE': 'Luxury', 'SHOPPING': 'Luxury'})
```

```python
grouped_task7
```

```python
grouped_task7=grouped_task7.groupby(['Type']).sum()
```

```python
#Calculating the most spent on category
sort_data3=grouped_task7.sort_values('Type', ascending=False)
Task7 = sort_data3.head(1)

Task7
```

```python
grouped_datanew3 = task8.groupby(['Year', 'Month']).agg({'Amount_x': 'sum','Amount_y': 'sum'})
```

```python
grouped_datanew3
```

```python
#Calculating debt and profit month wise, hence getting Monthly profit of the bank with interest rate of 2.9%
grouped_datanew3['Debt']=grouped_datanew3['Amount_x']-grouped_datanew3['Amount_y']
grouped_datanew3['Debt'] = grouped_datanew3['Debt'].apply(lambda x: max(0, x))  # Set negative surplus values to 0


grouped_datanew3['interest'] = grouped_datanew3['Debt'] * 0.029
grouped_datanew3 = grouped_datanew3.rename(columns={'interest': 'Monthly Profit'})
grouped_datanew3
```

```python
#Wrting all data to Excel file
with pd.ExcelWriter('output.xlsx') as writer:
 data.to_excel(writer, sheet_name='Age cleaned')
 groupT.to_excel(writer, sheet_name='Credit')
 Task1.to_excel(writer, sheet_name='Monthly Spend of each Customer')
 Task2.to_excel(writer, sheet_name='Monthly Repay of each Customer')
 Task3.to_excel(writer, sheet_name='Highest repaying 10 Customers')
 grouped_segment.to_excel(writer, sheet_name='Most spending segment')
 highest_age_group_sorted.to_excel(writer, sheet_name='Most spending Age group')
 sort_data2.to_excel(writer, sheet_name='Most profitable segment')
 sort_data3.to_excel(writer, sheet_name='Most Spend Category')
 grouped_datanew3.to_excel(writer, sheet_name='Monthly profit of bank')
```
